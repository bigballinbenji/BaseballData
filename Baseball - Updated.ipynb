```python
mlb = battingdata.merge(hallOfFame, left_on='playerID', right_on='playerID')
mlb.head(20)
```

```python
cols_obj = [cols for cols in mlb.columns if str(mlb[cols].dtype) =="object"]

cols_obj

mlb[cols_obj].describe().T
```

```python
playerGroup = mlb.groupby(['playerID'],as_index=False).mean()
playerGroup
```

```python
# Delete players who have null values for needed or votes
playerGroup.dropna(subset=["needed","votes"],inplace=True)
```

```python
playerGroup.isnull().sum()
```

```python
playerGroup.shape
```

```python
playerGroup['inducted'] = np.where(playerGroup['votes'] > playerGroup['needed'], 'Y', 'N')
```

```python
playerGroup
```

```python
playerGroup['inducted'].value_counts()
```

```python
mlb_gb = mlb.groupby('playerID', as_index=False).sum()
mlb_gb
mlb
```

```python
pg = playerGroup
```

```python
pg = pg.drop('2B', axis=1)
pg = pg.drop('3B', axis=1)
pg = pg.drop('CS', axis=1)
pg = pg.drop('SO', axis=1)
pg = pg.drop('IBB', axis=1)
pg = pg.drop('HBP', axis=1)
pg = pg.drop('SH', axis=1)
pg = pg.drop('SF', axis=1)
pg = pg.drop('GIDP', axis=1)
pg.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
#Splitting data
# Import train_test_split function


X=pg[['G', 'AB', 'R', 'H','HR','BB','RBI','SB','XB']]  # Features
y=pg['inducted']  # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

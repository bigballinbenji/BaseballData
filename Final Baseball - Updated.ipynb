## 1. Linear Regression

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Create your subset of features
baseball_X = playerGroup[["G","AB","RBI","R","XB","HR",'SO',"BB","SB"]] 
baseball_y = playerGroup["votes"]

```

```python
# Scale your data
scaler = StandardScaler()
scaler.fit(baseball_X,baseball_y)


#  Split the data and target into training/testing sets
baseball_X_train, baseball_X_test, baseball_y_train, baseball_y_test = train_test_split(baseball_X, baseball_y, test_size=0.2)

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(baseball_X_train, baseball_y_train)

# Make predictions using the testing set
baseball_y_pred = regr.predict(baseball_X_test)

```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
```

```python
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(baseball_y_test, baseball_y_pred))
```

```python
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(baseball_y_test, baseball_y_pred))
```

## 2. Logisitc Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sm
from sklearn.metrics import confusion_matrix

from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
# Create subset of features
X=playerGroup[['G', 'AB', 'RBI', 'XB', 'HR', 'SO', 'BB', 'SB']]
y=playerGroup['inducted']
```

```python
ros = RandomOverSampler(random_state=0)
X, y = ros.fit_resample(X, y)

print(sorted(Counter(y).items()))
```

```python
# Scale the data
scaler = StandardScaler()
scaler.fit(X,y)

#  Split the data and target into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Fit your logistic regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
# Print confusion matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

## 3. Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix

from imblearn.over_sampling import RandomOverSampler
```

```python
#Splitting data
# Import train_test_split function

X=pg[['G', 'AB', 'R', 'H','HR','BB','RBI','SB','XB']]  # Features
y=pg['inducted']  # Labels
```

```python
# Use this to solve for imbalance dataset

ros = RandomOverSampler(random_state=0)
X, y = ros.fit_resample(X, y)

print(sorted(Counter(y).items()))

```

```python

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Print confusion matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```
